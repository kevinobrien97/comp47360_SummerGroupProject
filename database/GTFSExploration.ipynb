<b>Actions</b>
- Account for calendar meanings on front end
- Merge stop_times and trips to get all of the stops for each route as well as the service_id. I will use this on the front end of the application using trip_headsign as the primary identifier, as well as service_id for the day. I will also obtain the route_short_name, in case the trip_headsign is duplicated for different routes (e.g. an express bus)
- Get the unique set of routes and their directions to replace the routes file

### Updating the stop_times file

```python
# merge stop_times and trips based on trip_id, setting the columns I want from trips
stop_times_updated = stop_times.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign']], on=['trip_id'], how='left')
```

```python
# removing columns from original stop_times file that are of no use
stop_times_updated = stop_times_updated.drop(columns=['pickup_type', 'drop_off_type', 'shape_dist_traveled', 'arrival_time'])
stop_times_updated
```

```python
# now merge stop_times and routes based on route_id, obtaining from it the route_short_name
stop_times_updated = stop_times_updated.merge(routes[['route_id', 'route_short_name']], on=['route_id'], how='left')
```
