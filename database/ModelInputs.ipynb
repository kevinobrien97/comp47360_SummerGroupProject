```python
import pandas as pd
import json 
```

In GTFSExploration.ipynb I created a file with all routes in each direction. As can be seen below however this cannot be directly merged with anything to obtain the direction_id. Therefore I have recreated the steps to create this, this time keeping direction_id. I will then store the trip_headsign and direction_id as a json object that can be looked up to find the direction_id.

```python
routes_updated = pd.read_csv('gtfs_files/routes_updated.txt')
```

```python
routes_updated
```

```python
trips = pd.read_csv('gtfs_files/trips.txt')
```

```python
trips
```

```python
# merge stop_times and trips based on trip_id, setting the columns I want from trips
headsign_direction = routes_updated.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign']], on=['trip_id'], how='left')
```

```python
stop_times = pd.read_csv('gtfs_files/stop_times.txt')
```

```python
# merge stop_times and trips based on trip_id, setting the columns I want from trips
stop_times_updated = stop_times.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign', 'direction_id']], on=['trip_id'], how='left')
```

```python
# removing columns from original stop_times file that are of no use
stop_times_updated = stop_times_updated.drop(columns=['pickup_type', 'drop_off_type', 'shape_dist_traveled', 'arrival_time'])
stop_times_updated
```

```python
routes = pd.read_csv('gtfs_files/routes.txt')
```

```python
# now merge stop_times and routes based on route_id, obtaining from it the route_short_name
stop_times_updated = stop_times_updated.merge(routes[['route_id', 'route_short_name']], on=['route_id'], how='left')
```

```python
stop_times_updated
```

```python
route_direction = stop_times_updated[['trip_headsign', 'route_short_name', 'direction_id']].drop_duplicates()
route_direction = route_direction.reset_index(drop=True)
```

```python
route_direction
```

```python
# create empty object to store direction IDs
direction_id = {}
```

```python
for index, row in route_direction.iterrows():
    direction_id[row["trip_headsign"]] = str(row['direction_id'])
    
```

```python
direction_id
```

```python
json_object = json.dumps(direction_id)
```

```python
with open('../dubbus/models/direction_id.json', 'w') as outfile:
    outfile.write(json_object)
```
