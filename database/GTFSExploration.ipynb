### Updating the stop_times file

```python
# merge stop_times and trips based on trip_id, setting the columns I want from trips
stop_times = stop_times.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign']], on=['trip_id'], how='left')
```

```python
# removing columns from original stop_times file that are of no use
stop_times = stop_times.drop(columns=['pickup_type', 'drop_off_type', 'shape_dist_traveled'])
stop_times
```

```python
stop_times[['trip_headsign', 'route_id']].describe().T
```

```python
# export to csv
stop_times.to_csv('stop_times_updated.txt', index=False)
```

### Getting unique routes in each direction

```python
# all unique combinations of trip_headsign and route_id from stop_times
route_headsign = stop_times[['trip_headsign', 'route_id']].drop_duplicates()
route_headsign = route_headsign.reset_index(drop=True)
```

```python
route_headsign
```

```python
# join on routes to get the short_name
route_headsign = routes.join(route_headsign.set_index('route_id'), on='route_id')
```

```python
route_headsign
```

```python
# route_headsign = route_headsign[route_headsign.trip_headsign.notna()]
route_headsign = route_headsign.reset_index(drop=True)
```

```python
route_headsign
```

```python
route_headsign = route_headsign.drop(columns=['route_type','agency_id','route_long_name'])
```

```python
route_headsign.describe()
```

```python
route_headsign.to_csv('routes_updated.txt', index=False)
```

### Getting unique stops for each route_headsign

```python
# getting the subset of rows that are unique based on the 'stop_id', 'trip_headsign', 'route_id'
stops_route_headsign = stop_times[['stop_id', 'trip_headsign', 'route_id']].drop_duplicates()
```

```python
stops_route_headsign = stops_route_headsign.reset_index(drop=True)
```

```python
stops_route_headsign
```

```python
# testing if stop_ids listed more than once (they should as a lot of stops are serviced by multiple routes)
stops_route_headsign['stop_id'].describe()
```

```python
# merge stops_route_headsign and stop based on stop_id
stops_route_headsign = stops_route_headsign.merge(stops, on=['stop_id'], how='left')
```
