## JSON Object to map trip_headsign to direction_id

In GTFSExploration.ipynb I created a file with all routes in each direction. As can be seen below however this cannot be directly merged with anything to obtain the direction_id. Therefore I have recreated the steps to create this, this time keeping direction_id. I will then store the trip_headsign and direction_id as a json object that can be looked up to find the direction_id.

```python
routes_updated = pd.read_csv('gtfs_files/routes_updated.txt')
```

```python
routes_updated
```

```python
trips = pd.read_csv('gtfs_files/trips.txt')
```

```python
trips
```

```python
# # merge stop_times and trips based on trip_id, setting the columns I want from trips
# headsign_direction = routes_updated.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign']], on=['trip_id'], how='left')
```

```python
stop_times = pd.read_csv('gtfs_files/stop_times.txt')
```

```python
# merge stop_times and trips based on trip_id, setting the columns I want from trips
stop_times_updated = stop_times.merge(trips[['trip_id', 'route_id', 'service_id','trip_headsign', 'direction_id']], on=['trip_id'], how='left')
```

```python
# removing columns from original stop_times file that are of no use
stop_times_updated = stop_times_updated.drop(columns=['pickup_type', 'drop_off_type', 'shape_dist_traveled', 'arrival_time'])
stop_times_updated
```

```python
stop_times_updated['stop_headsign'].describe()
```

```python
routes = pd.read_csv('gtfs_files/routes.txt')
```

```python
# now merge stop_times and routes based on route_id, obtaining from it the route_short_name
stop_times_updated = stop_times_updated.merge(routes[['route_id', 'route_short_name']], on=['route_id'], how='left')
```

```python
stop_times_updated
```

```python
route_direction = stop_times_updated[['trip_headsign', 'route_short_name', 'direction_id']].drop_duplicates()
route_direction = route_direction.reset_index(drop=True)
```

```python
route_direction
```

```python
# create empty object to store direction IDs
direction_id = {}
```

```python
for index, row in route_direction.iterrows():
    # the equivalent direction IDs are 1 and 2 in the dataset we used to train our models 
    if row['direction_id'] == 0:
        direction_id[row["trip_headsign"]] = "1"
    else:
        direction_id[row["trip_headsign"]] = "2"
```

```python
direction_id
```

```python
json_object = json.dumps(direction_id)
```

```python
with open('../dubbus/models/direction_id.json', 'w') as outfile:
    outfile.write(json_object)
```

## JSON Object to map stop_name to stop_sequence (i.e. progress number)

First I will create a similar dataframe to the one above, but this time creating a column combining the route name and the direction. This is how our pickle files are named and will allow for easy use of the progress_numbers object as I will use it as a key for an object containing all of the stop_names for that route.

As our 2018 dataset has different mappings for directions (0=1 and 1=2), I will first make a new column with this change

```python
route_direction['direction_id_updated'] = route_direction.apply(lambda row: 1 if row["direction_id"]==0 else 2, axis=1)
```

```python
route_direction['route_id'] = route_direction.apply(lambda row: row["route_short_name"] + "_" + str(row["direction_id_updated"]), axis=1)
```

```python
route_direction
```

From previous work I created a dataframe with the stop_id, headsign and stop_sequence. I will recreate this now.

```python
# getting the subset of rows that are unique based on the 'stop_id', 'trip_headsign', 'route_short_name'
stops_route_headsign = stop_times_updated[['stop_id', 'trip_headsign', 'route_short_name', 'stop_sequence']].drop_duplicates()
```

```python
stops_route_headsign = stops_route_headsign.reset_index(drop=True)
```

```python
stops_route_headsign
```

Above is a dataframe with the stops in each route in the Dublin bus network, alongside the sequence they are served in each route. Now I will need to map the stop_id to the name, as the name is what we receive from Google Maps.

```python
stops = pd.read_csv('gtfs_files/stops.txt')
```

```python
stops
```

```python
# merge to get stop_name
stops_route_headsign = stops_route_headsign.merge(stops[['stop_id', 'stop_name']], on=['stop_id'], how='left')
```

```python
stops_route_headsign
```

Now I need to create a JSON object that contains the stop names for each route directions with their progress number

```python
# empty object
progress_numbers = {}
```

```python
# iterate through all of the routes - each is going to be a key, with its value being another object containing all of the stops
for index, row in route_direction.iterrows():
    # temporary dataframe with stops from stops_routes_headsign matching those in the current row
    temp_df = stops_route_headsign.loc[stops_route_headsign['trip_headsign']==row['trip_headsign']]
    # define a temporary object
    temp_obj = {}
    
    # below iterates through the temp_df, adding stop names and IDs
    for index1, row1 in temp_df.iterrows():
        temp_obj[row1["stop_name"]] = str(row1['stop_sequence'])
        
    # create new dict in progress_numbers named after the route and direction e.g. 46A_1 with its value being temp_obj
    progress_numbers[row['route_id']] = temp_obj 
```

```python
progress_numbers
```

```python
json_object1 = json.dumps(progress_numbers)
```

```python
with open('../dubbus/models/progress_numbers.json', 'w') as outfile:
    outfile.write(json_object1)
```
